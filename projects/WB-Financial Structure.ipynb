```python
import matplotlib.ticker as ticker

ll = apac.xs(2017, level='year').iloc[:,1].dropna()
#ll_17_avg = pd.Series(data=[ll_17.mean()] * 10, index=ll_17.index)
db = apac.xs(2017, level='year').iloc[:,3].dropna()

fig, axes = plt.subplots(ncols = 2, sharey = True, figsize=(15,7))

axes[0].barh(db.index, db, align='center')
axes[1].barh(ll.index, ll, align='center')
axes[0].set(title=db.name)
axes[1].set(title=ll.name)

axes[0].invert_xaxis()
axes[0].set(yticklabels=db.index)
axes[0].yaxis.tick_right()
axes[0].set_xlim(left=200)

axes[0].xaxis.set_major_locator(ticker.MultipleLocator(50))
axes[1].xaxis.set_major_locator(ticker.MultipleLocator(50))
axes[0].xaxis.set_minor_locator(ticker.AutoMinorLocator())
axes[1].xaxis.set_minor_locator(ticker.AutoMinorLocator())

#plt.scatter(ll_17_avg.index, ll_17_avg, color='r')

axes[0].spines['left'].set_visible(False)
axes[0].spines['top'].set_visible(False)
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)   

plt.tight_layout()
fig.subplots_adjust(wspace=0.1)
plt.show()
```
